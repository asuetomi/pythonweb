### HTML基礎
~~~html
<!DOCTYPE html>
<html>
 <head></head>
 <body></body>
</html>
~~~
- リンクタグ〈別ページへのリンク〉
- リンクタグ〈同一ペ寺内リンク）
- 画像ファイルの表示
- テーブルタグ table tr th td
- テーブルタグ
- 段落 p
- ブロックレベルでグループ化 div
- インラインレベルでグループ化 span
- 入力の受付には form タグを使用する
~~~html
<form action="program.file" method="post"></form>
~~~
post と get。  
get は url に表示される。
post は url に含まれないのでブックマークしても復元できない。

- テキストの入力を行う input タグ
~~~html
<input type="text" name="name">
~~~
- input タグの type を radio に変更することでラジオボタン
~~~html
<input type="radio" name="name">
~~~
- セレクトボックス select
- 複数行入力可能なテキストエリア
~~~html
<input type="textarea" name="name">
~~~
- 送信ボタン
~~~html
<input type="submit" value="送信">
~~~


### DJANGOを導入する
導入する
~~~console
pip install django
~~~
プロジェクトを作成する
~~~console
django-admin startproject pythonweb
~~~
作成したプロジェクトに移動
~~~console
cd pythonweb
~~~
アプリケーションを作成します
~~~console
python manage.py startapp myapp
~~~
実行してみましょう
~~~console
python manage.py runserver
~~~
下記にアクセスして表示
~~~url
http://localhost:8000/
~~~

Webから値を取得してみよう
\myapp に forms.py というファイルを新規作成
~~~
from django import forms

class TestForm(forms.Form):
    txt = forms.CharField(label='文字')
    num = forms.IntegerField(label='数値')
~~~
forms.html
~~~html
<html>
    <head>
        <title>Form test</title>
    </html
    <body>
        <p>{{insert_data|safe}}</p>
        <form action="{% url 'formtest' %}" method='POST'>
            {% csrf_token %}
            {{ form }}
            <input type="submit" value="送信">
        </form>
    </body>
</html>
~~~
